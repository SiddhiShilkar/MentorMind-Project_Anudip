UNIVARIATE ANALYSIS

1. Examine Distributions of Individual Variables: Plot histograms or bar charts to visualize the distribution of key variables such as age, BMI, mental health days (Menthlth), and physical health days (PhysHlth).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('C:/Users/Vilas/Downloads/cleaned_dataset.xlsx')
```

Histogram for Age:

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=15, color='lightblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Age appears to be normally distributed, with most individuals falling within the range of 5 to 11. This suggests a fairly balanced dataset in terms of age.

Histogram for BMI:

```python
plt.figure(figsize=(8, 6))
plt.hist(df['BMI'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

The BMI is skewed towards lower values, with a majority of individuals having a BMI between 20 and 30. Few individuals have extreme BMI values (either too low or too high).

Histogram for Mental Health Days (MentHlth):

```python
plt.figure(figsize=(8, 6))
plt.hist(df['MentHlth'], bins=15, color='salmon', edgecolor='black')
plt.title('Distribution of Mental Health Days')
plt.xlabel('Number of Mental Health Days')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Most individuals report having very few or no mental health days, with the data being highly skewed towards zero.

Histogram for Physical Health Days (PhysHlth):

```python
plt.figure(figsize=(8, 6))
plt.hist(df['PhysHlth'], bins=15, color='orange', edgecolor='black')
plt.title('Distribution of Physical Health Days')
plt.xlabel('Number of Physical Health Days')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Similar to mental health days, most individuals report having fewer physical health issues, with a skew towards zero days.

2. Investigate Prevalence of Health Conditions: Plot bar charts or pie charts to show the prevalence of high blood pressure (HighBP), high cholesterol (HighChol), and other health conditions.

Bar chart for High Blood Pressure (HighBP):

```python
plt.figure(figsize=(6, 6))
df['HighBP'].value_counts().plot(kind='bar', color=['lightblue', 'salmon'], edgecolor='black')
plt.title('Prevalence of High Blood Pressure')
plt.xlabel('High Blood Pressure')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

The dataset shows that a significant portion of the population suffers from high blood pressure, though those without high blood pressure outnumber those with it.

Pie chart for High Cholesterol (HighChol):

```python
plt.figure(figsize=(6, 6))
df['HighChol'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Prevalence of High Cholesterol')
plt.ylabel('')
plt.show()
```

42.2% of individuals have high cholesterol, while 57.8% do not. There’s a sizable population affected by cholesterol-related issues.

Bar chart for Cholesterol Check (CholCheck):

```python
plt.figure(figsize=(6, 6))
df['CholCheck'].value_counts().plot(kind='bar', color=['lightgreen', 'orange'], edgecolor='black')
plt.title('Prevalence of Cholesterol Check')
plt.xlabel('Cholesterol Check')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

The majority of the population has undergone a cholesterol check, indicating high health awareness in this aspect.

3. Analyze Distribution of Heart Disease (Target Variable): Visualize the distribution of heart disease cases using a bar chart or pie chart.

Bar chart for Heart Disease (Target Variable):

```python
plt.figure(figsize=(6, 6))
df['HeartDiseaseorAttack'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'], edgecolor='black')
plt.title('Distribution of Heart Disease Cases')
plt.xlabel('Heart Disease or Attack')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

The vast majority of individuals do not have heart disease (as indicated by the much taller blue bar). A significantly smaller proportion of individuals suffer from heart disease, as indicated by the shorter red bar.

Pie chart for Heart Disease (Target Variable):

```python
plt.figure(figsize=(6, 6))
df['HeartDiseaseorAttack'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Heart Disease Distribution')
plt.ylabel('')
plt.show()
```

The dataset shows that 10.3% of the population has heart disease, while the vast majority (89.7%) do not, which is a positive indication of overall heart health.

BIVARIATE ANALYSIS

```python
import seaborn as sns
```

1. Explore Relationships with Heart Disease: Compare the distribution of key variables (e.g., HighBP, HighChol, BMI) between individuals with and without heart disease.

Bar Chart for High Blood Pressure vs Heart Disease

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='HighBP', hue='HeartDiseaseorAttack', data=df, palette='Set2')
plt.title('High Blood Pressure vs Heart Disease')
plt.xlabel('High Blood Pressure')
plt.ylabel('Count')
plt.show()
```

 Those with heart disease are more likely to have high blood pressure compared to those without heart disease. This suggests a strong relationship between high blood pressure and heart disease.

Boxplot for BMI vs Heart Disease

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=df, palette='Set2')
plt.title('BMI Distribution by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('BMI')
plt.show()
```

The BMI distribution for those with heart disease appears similar to those without heart disease, though individuals with heart disease might have slightly higher BMI on average.

2. Visualize Correlations Between Variables: Use scatter plots or correlation matrices to explore relationships between continuous variables.

```python
plt.figure(figsize=(10, 8))
corr = df[['BMI', 'Age', 'MentHlth', 'PhysHlth']].corr()  # Select continuous variables
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Continuous Variables')
plt.show()
```

1. BMI and physical health days show a weak positive correlation (0.12).

2. Age and BMI have a slight negative correlation (-0.037), indicating that BMI tends to slightly decrease with age.

3. There is a stronger correlation (0.35) between BMI and physical health days, suggesting that individuals with higher BMI may experience more physical health issues.

```python
# Scatter Plot - Age vs BMI with legend placement
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='BMI', hue='HeartDiseaseorAttack', data=df, palette='Set2')
plt.title('Age vs BMI colored by Heart Disease')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(loc='upper right')  # Specify the location
plt.show()
```

3. Compare Heart Disease Across Demographic Groups: Analyze the prevalence of heart disease across different demographic groups (e.g., age, gender, education) using bar charts or pivot tables.

Bar Chart for Heart Disease across Gender

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='HeartDiseaseorAttack', data=df, palette='Set2')
plt.title('Heart Disease Prevalence across Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

The majority of individuals, irrespective of gender, do not suffer from heart disease (as indicated by the taller green bars). The orange bars, representing individuals with heart disease, are much smaller for both genders. However, it appears that one gender has a higher overall prevalence of heart disease.

Pivot Table for Heart Disease across Education Levels

```python
pivot_edu = df.pivot_table(values='HeartDiseaseorAttack', index='Education', aggfunc='count')
print(pivot_edu)
```

As the education level increases from 1 to 6, the number of individuals increases significantly. This suggests that higher education levels have a larger representation in the dataset. The data can also be used to compare the prevalence of heart disease at different education levels by referring to the detailed count of heart disease cases.

Bar Chart for Heart Disease across Education Levels

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Education', hue='HeartDiseaseorAttack', data=df, palette='Set2')
plt.title('Heart Disease Prevalence across Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

As education levels increase, the overall number of individuals (both with and without heart disease) increases as well. However, the proportion of individuals with heart disease (orange bars) is relatively low at all education levels, with a slight increase in the later stages (levels 4, 5, and 6).
